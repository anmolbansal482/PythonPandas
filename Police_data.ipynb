```python
import pandas as pd
```

```python
df = pd.read_csv(r'Police Data.csv')
```

```python
df
```

```python
#finding null values of each column
df.isnull().sum()
```

```python
#Removing column with null values (temporarily)
df.drop(columns = 'country_name')
```

```python
#Removing a column permanently using inplace function
df.drop(columns = 'country_name', inplace = True)
```

```python
df
```

```python
df.head(5)
```

```python
#filtering data to find a specific pattern
df[df.violation == 'Speeding']
df.head(5)
```

```python
#finding a specific pattern along with the numbers of counts of different columns
df[df.violation == 'Speeding'].driver_gender.value_counts()
```

```python
#grouping the gender column to find out the number of times each gender was stopped for search
df.groupby('driver_gender').search_conducted.sum()
```

```python
#finding the mean of the stop duration and changing it's data type to be able to find the mean from str to int
dtyp = df.dtypes['stop_duration']
print(dtyp)
```

```python
#mapping and type casting to find the mean of a column with object data type and string values
df['stop_duration'] = df['stop_duration'].map({'0-15 Min' : 7.5, '16-30 Min' : 24, '30+ Min' : 45 }, 
                                              na_action = 'ignore')
```

```python
df['stop_duration'].mean()
```

```python
#using group by and describe function to find violation with age groups
df.groupby('driver_age').violation.describe().head(50)
```
